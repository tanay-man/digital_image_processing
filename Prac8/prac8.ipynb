```python
import numpy as np 
```

```python
I = np.array([[154,123,123,123,123,123,123,136],
              [192,180,136,154,154,154,136,110],
              [254,198,154,154,180,154,123,123],
              [239,180,136,180,180,166,123,123],
              [180,154,136,167,166,149,136,136],
              [128,136,123,136,154,180,198,154],
              [123,105,110,149,136,136,180,166],
              [110,136,123,123,123,136,154,136]])
```

```python
M = I -128
```

```python
T = np.zeros((8,8))
for i in range(8):
    for j in range(8):
        if i == 0:
            T[i,j] = np.sqrt(1/8)
        else:
            T[i,j] = np.sqrt(1/4) * np.cos((2*j + 1)*i*np.pi/16)
```

```python
D = np.dot(np.dot(T,M),np.transpose(T))
```

```python
Q = np.array([
    [16, 11, 10, 16, 24, 40, 51, 61],
    [12, 12, 14, 19, 26, 58, 60, 55],
    [14, 13, 16, 24, 40, 57, 69, 56],
    [14, 17, 22, 29, 51, 87, 80, 62],
    [18, 22, 37, 56, 68, 109, 103, 77],
    [24, 35, 55, 64, 81, 104, 113, 92],
    [49, 64, 78, 87, 103, 121, 120, 101],
    [72, 92, 95, 98, 112, 100, 103, 99]
])
```

```python
C = np.round(D/Q)
R = Q * C
N = np.round(np.dot(np.dot(np.transpose(T),R),T)+128)
```

```python
print("M",M)
print()
print("Q",Q)
print()
print("D",D)
print()
print("C",C)
print()
print("R",R)
print()
print("I",I)
print()
print("N",N)
```
