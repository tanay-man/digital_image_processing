```python
import cv2
import numpy as np 
from matplotlib import pyplot as plt 
image_path = r"../images/"
```

```python
#Loads and display image
sample_image = cv2.imread(image_path+"1.jpg",1)
window_name = "sample"
cv2.imshow(window_name,sample_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Resize Image
small = cv2.resize(sample_image,(2000,1000))
cv2.imshow("sample",small)
cv2.waitKey(0)
```

```python
# Rotate
clock = cv2.rotate(small,cv2.ROTATE_90_CLOCKWISE)
anti = cv2.rotate(small,cv2.ROTATE_90_COUNTERCLOCKWISE)
upside = cv2.rotate(small,cv2.ROTATE_180)
cv2.imshow("Clockwise",clock)
cv2.waitKey(0)
cv2.imshow("AntiClockWise",anti)
cv2.waitKey(0)
cv2.imshow("UpsideDown",upside)
cv2.waitKey(0)
```

```python
# Add images
img2 = cv2.imread(image_path+"2.jpg",1)
small2 = cv2.resize(img2,(2000,1000))
sum = cv2.add(small,small2)
cv2.imshow("Added image",sum)
cv2.waitKey(0)
```

```python
# Plot histogram
r_hist = cv2.calcHist(sample_image,channels=[0],mask=None,histSize=[256],ranges = [0,256])
plt.plot(r_hist, color='r') 
plt.title('Image Histogram For Red channel') 
plt.xlabel("Value of pixels")
plt.ylabel("Number of pixels")
plt.show()
```

```python
# Colour conversion
grey_scale = cv2.cvtColor(small,cv2.COLOR_BGR2GRAY)
cv2.imshow("gray scale",grey_scale)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Thresholding
thresh = cv2.threshold(grey_scale, 100, 256,cv2.THRESH_BINARY)
cv2.imshow("Thresholding",thresh[1])
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Noise
mean = (128,128,128)
sigma = (150,150,150)
noisy_img = cv2.randn(small,mean,sigma)
cv2.imshow("Nosiy Image",noisy_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
